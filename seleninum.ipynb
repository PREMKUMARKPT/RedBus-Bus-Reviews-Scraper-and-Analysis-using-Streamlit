```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

element_list = []

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

try:
    for page in range(1, 3):
        page_url = f"https://www.redbus.in/search?fromCityName=CMBT%2C%20Chennai%2C%20Chennai&fromCityId=66065&srcCountry=IND&fromCityType=AREA&toCityName=Bangalore&toCityId=122&destCountry=India&toCityType=CITY&onward=26-Jun-2025&doj=26-Jun-2025&ref=modifySearch&page={page}"
        driver.get(page_url)

        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "travelsName___950ec8"))
        )

        title = driver.find_elements(By.CLASS_NAME, "travelsName___950ec8")
        price = driver.find_elements(By.CLASS_NAME, "finalFare___057afc")
        description = driver.find_elements(By.CLASS_NAME, "busType___675d0a")
        rating = driver.find_elements(By.CLASS_NAME, "rating___f7ba20")

        for i in range(len(title)):
            element_list.append([title[i].text, price[i].text, description[i].text, rating[i].text])

finally:
    driver.quit()

print(element_list)
```

```python
import xlsxwriter
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

element_list = []

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

try:
    # Only one page, pagination might be dynamic
    driver.get("https://www.redbus.in/search?fromCityName=CMBT%2C%20Chennai%2C%20Chennai&fromCityId=66065&srcCountry=IND&fromCityType=AREA&toCityName=Thiruvannamalai&toCityId=293&destCountry=IND&toCityType=CITY&onward=26-Jun-2025&doj=26-Jun-2025&ref=home")

    WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, "travelsName___950ec8"))
    )

    title = driver.find_elements(By.CLASS_NAME, "travelsName___950ec8")
    price = driver.find_elements(By.CLASS_NAME, "finalFare___057afc")
    description = driver.find_elements(By.CLASS_NAME, "busType___675d0a")
    rating = driver.find_elements(By.CLASS_NAME, "rating___f7ba20")

    for i in range(len(title)):
        element_list.append([title[i].text, price[i].text, description[i].text, rating[i].text])

finally:
    driver.quit()

# Writing to Excel
workbook = xlsxwriter.Workbook("redbus_data.xlsx")
worksheet = workbook.add_worksheet()

headers = ["Title", "Price", "Description", "Rating"]
for col, header in enumerate(headers):
    worksheet.write(0, col, header)

for row, item in enumerate(element_list, start=1):
    for col, value in enumerate(item):
        worksheet.write(row, col, value)

workbook.close()
```
